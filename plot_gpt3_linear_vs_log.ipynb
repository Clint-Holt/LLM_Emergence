```python
# Imports
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.size'] = '20'
plt.rcParams['axes.labelsize'] = '20'
plt.rcParams['ytick.labelsize'] = '20'
```

```python
# Load Data
data = """
GPT-3_model params(B)  performance	error_rate
Small	    0.125	    0.5	        99.5
Medium	    0.35	    0.4	        99.6
Large	    0.76	    0.4	        99.6
XL	        1.3	        0.2	        99.8
3B	        2.7	        0.6	        99.4
6B	        6.7	        1.4	        98.6
13B	        13	        7.1	        92.9
175B	    175	        32.7	    67.3
"""
params = [0.125, 0.35, 0.76, 1.3, 2.7, 6.7, 13.0, 175.0]
performance = [0.5, 0.4, 0.4, 0.2, 0.6, 1.4, 7.1, 32.7]
error_rate = [99.5, 99.6, 99.6, 99.8, 99.4, 98.6, 92.9, 67.3]
# Source: https://github.com/google/BIG-bench/tree/main/bigbench/benchmark_tasks/modified_arithmetic/results
```

```python
plt.title("GPT-3 Model Performance (linear scale)")
plt.ylabel("Performance")
plt.xlabel("Billions of Parameters (linear scale)")
plt.plot(params, performance)
```

```python
plt.title("GPT-3 Model Performance (log scale)")
plt.ylabel("Performance")
plt.xlabel("Billions of Parameters (log scale)")
plt.semilogx(params, performance, marker='o')
```

```python
plt.title("GPT-3 Error Rate (LinearX/LogY scale)")
plt.ylabel("Performance")
plt.xlabel("Billions of Parameters (log scale)")
plt.semilogx(params, error_rate, marker='o')
```

```python
plt.title("GPT-3 Error Rate (LogX/LogY scale)")
plt.ylabel("Performance")
plt.xlabel("Billions of Parameters (log scale)")
plt.loglog(params, error_rate, marker='o')
```

```python
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.title("GPT-3 Model Performance (linear scale)")
plt.ylabel("Performance")
plt.xlabel("Billions of Parameters (linear scale)")
plt.plot(params, performance)

plt.subplot(2, 2, 2)
plt.title("GPT-3 Model Performance (log scale)")
plt.ylabel("Performance")
plt.xlabel("Billions of Parameters (log scale)")
plt.semilogx(params, performance, marker='o')

plt.subplot(2, 2, 3)
plt.title("GPT-3 Error Rate (LinearX/LogY scale)")
plt.ylabel("Performance")
plt.xlabel("Billions of Parameters (log scale)")
plt.semilogx(params, error_rate, marker='o')

plt.subplot(2, 2, 4)
plt.title("GPT-3 Error Rate (LogX/LogY scale)")
plt.ylabel("Performance")
plt.xlabel("Billions of Parameters (log scale)")
plt.loglog(params, error_rate, marker='o')

plt.tight_layout()
plt.savefig("tmp.png", dpi=300)
```
